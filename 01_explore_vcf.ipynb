## Reading cleaned VCF

```python
from cyvcf2 import VCF
import pandas as pd 

vcf = VCF('data/cleaned.vcf')

rows = []
for record in vcf:
    info = record.INFO
    rows.append({
        'Chromosome': record.CHROM,
        'Position': record.POS,
        'Reference': record.REF,
        'Alternate_Allele': ",".join(map(str, record.ALT)),
        "Gene_Symbol": info.get("GENEINFO", "").split(":")[0] if info.get("GENEINFO") else None,
        "Clinical_Significance": info.get("CLNSIG"),
        "Clinical_Review_Status": info.get("CLNREVSTAT"),
        "Clinical_Disease_Name": info.get("CLNDN"),
        "Missense_Variant": info.get("MC"),
        "Allele_Frequencies_ESP": info.get("AF_ESP"),
        "Allele_Frequencies_EXAC": info.get("AF_EXAC"),
        "Allele_Frequencies_TGP": info.get("AF_TGP")
    })
    
df = pd.DataFrame(rows)

# Save clean, readable TSV
df.to_csv("data/genomics_clean.tsv", sep="\t", index=False)
```
